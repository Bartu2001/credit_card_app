```python
def luhn_algorithm(card_number):
    card_number = card_number[::-1]
    total = 0
    for i in range (len(card_number)):
        digit =int(card_number[i])
        
        if i % 2 ==1:
            digit *= 2
            if digit > 9:
                digit -= 9

        total += digit
        

    if total % 10 == 0:
        return True
    else:
        return False
              
          
    
def kart_tipi(card_number):
    

    if len(card_number) == 16:
        if card_number[0] == "4":
            return "Visa"
        elif card_number[0] =="5":
            return "MasterCard"
        else:
            return "Geçersiz Kart"
    elif len(card_number) == 15 and (card_number[:2] =="34" or card_number[:2] =="37"):
        return "American Express"
    else:
        return "Geçersiz kart"



kart_numarasi = input("Kredi Kartı Numarası Girin: ").replace(" ", "")



if len(kart_numarasi) !=15 and len(kart_numarasi) != 16:
    print("Geçersiz kart numarası uzunluğu")
else:
    kart_tip =kart_tipi(kart_numarasi)
    
    if kart_tip !="Geçersiz kart":
        if luhn_algorithm(kart_numarasi):
            print(f"Girdiğiniz kredi kartı geçerli bir {kart_tip} kartıdır.")
        else:
            print(f"Girdiğiniz kredi kartı geçersizdir.")
    else:
        print("Geçersiz kart tipi")


```
